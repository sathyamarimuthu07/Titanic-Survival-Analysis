```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Titanic-Dataset.xlsx")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['Survived'].value_counts()
```

```python
df.groupby('Sex')['Survived'].mean()
```

```python
df.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Gender")
plt.show()
```

```python
df.groupby('Pclass')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
df['Age'].hist()
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
df['Survived'].value_counts().plot(kind='bar')
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
df.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
df.groupby('Pclass')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
df['Age'].hist(bins=20)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
df.groupby('Embarked')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Embarkation Port")
plt.xlabel("Embarked Port")
plt.ylabel("Survival Rate")
plt.show()
```

```python
df.groupby('Pclass')['Fare'].mean().plot(kind='bar')
plt.title("Average Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Fare")
plt.show()
```

```python
df[['Survived','Pclass','Age','Fare']].corr()
```

```python
import seaborn as sns

sns.heatmap(df[['Survived','Pclass','Age','Fare']].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
